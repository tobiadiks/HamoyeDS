```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jul 31 18:06:21 2020

@author: USER
"""

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso
df=pd.read_csv("EDC.csv")
df=df.drop(columns=["date","lights"])
scaler=MinMaxScaler()
normalized_df=pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
target=normalized_df["Appliances"]
#target is for the appliances

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(normalized_df,target,test_size=0.3,random_state=42)
#linear model
linear_reg=LinearRegression()
linear_reg=linear_reg.fit(x_train,y_train)
#end of linear model

ridge=Ridge()
ridge=ridge.fit(x_train, y_train)
#end ridge
lasso=Lasso()
lasso=lasso.fit(x_train,y_train)


def get_weight(model, feat, col_name):
    weight=pd.Series(model.coef_, feat.columns).sort_values()
    weight_df=pd.DataFrame(weight).reset_index()
    weight_df.columns=["Features",col_name]
    weight_df[col_name].round(3)
    return weight_df

linear_weight=get_weight(linear_reg,x_test,"Linear-Regression-Weight")
lasso_weight=get_weight(lasso,x_test,"Lasso-weight")
ridge_weight=get_weight(ridge,x_test,"Ridge-weight")
                         
final_weight=pd.merge(linear_weight,lasso_weight,on="Features")
final_weight=pd.merge(final_weight,ridge_weight,on="Features")

from sklearn.metrics import r2_score,mean_absolute_error
def feat_metrics(predicted):
    pred=predicted.predict(x_test)
    r2=r2_score( y_test,pred)
    mae=mean_absolute_error(y_test,pred)
    r2,mae=round(r2,3),round(mae,3)
    return r2,mae



  #endpoint  
    
    
    
    
print(final_weight)
print(feat_metrics(linear_reg))
print(feat_metrics(ridge))
print (feat_metrics(lasso))
```
